```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
```

```python
df = pd.read_csv("breast-cancer.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
# to find null value
df.isnull().sum()
```

```python
df.drop(['Unnamed: 32'], axis = 1 ,inplace=True)
```

```python
df
```

```python
df.shape
```

```python
df['diagnosis'].value_counts()
```

```python
sns.countplot(df['diagnosis'],label='count')
plt.show()
```

```python
df.dtypes         # only one is object
```

```python
df['diagnosis']=df['diagnosis'].map({'B':0,'M':1})
```

```python
df.head()
```

```python
x = df.iloc[:,2:].values
y = df.iloc[:,1].values
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y, test_size=.2, random_state=12)
```

```python
from sklearn.preprocessing import StandardScaler

std = StandardScaler()
```

```python
std_train = std.fit_transform(x_train)
std_test = std.transform(x_test)
```

```python
std_train.shape
```

```python
std_test.shape
```

### Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(criterion = 'gini', random_state = 12)

dt.fit(std_train,y_train)
```

```python
dt_train_score = dt.score(std_train,y_train)

dt_train_score
```

```python
dt_test_score = dt.score(std_test,y_test)

dt_test_score
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,dt.predict(std_test))
```

### Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators = 10 ,random_state=12)

rf.fit(std_train,y_train)
```

```python
rf_train_score = rf.score(std_train,y_train)

rf_train_score
```

```python
rf_test_score = rf.score(std_test,y_test)

rf_test_score
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,rf.predict(std_test))
```

### SVM

```python
from sklearn.svm import SVC
```

```python
clf = SVC(kernel = 'linear',C=10)

clf.fit(std_train,y_train)
```

```python
clf_train_score = clf.score(std_train,y_train)

clf_train_score
```

```python
clf_test_score = clf.score(std_test,y_test)

clf_test_score
```

```python
confusion_matrix(y_test,clf.predict(std_test))
```

```python
log = LogisticRegression(random_state=12)

log.fit(std_train,y_train)
```
